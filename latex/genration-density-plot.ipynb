```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MaxNLocator
import numpy as np
```

```python
ns = np.arange(1, 4+1)

def dist(k: int):
    def p(n: int):
        power = 1 / (k - 1)
        return (n / 4) ** power - ((n - 1) / 4) ** power
    return p

ns
```

```python
rc_context = {
    'text.usetex': True,
    'font.family': 'Helvetica',
}

with plt.rc_context(rc=rc_context):
    fig, ax = plt.subplots()

    for k in [2, 3, 4]:
        vals = dist(k)(ns)
        ax.scatter(ns, vals, label=f'$|L|={k}$')
        ax.plot(ns, vals, linestyle='--')

    ax.legend(prop={'size': 16})
    ax.set_ylim(bottom=0)
    ax.xaxis.set_major_locator(MaxNLocator(integer=True))
    ax.set_ylabel('$p(f_l)$', size=16)
    ax.set_xlabel('$f_l$', size=16)

    fig.savefig('figures/generation-rl-sample-density.pdf')
```
