```python
import networkx as nx
import matplotlib.pyplot as plt

# Generate a 3x3 grid graph
G = nx.grid_graph(dim=[5, 5])

# Draw the graph
nx.draw(G, with_labels=True)
plt.show()
```

```python
# We can get the adjacency matrix of G as a numpy array using the adjacency_matrix() method of networkx graph object.
# Here's the code to get the adjacency matrix of G as a numpy array:

adj_matrix = nx.adjacency_matrix(G).toarray()
print(adj_matrix)
```

```python
# Get the degree matrix of G as a numpy array
degree_matrix = np.diag(np.sum(adj_matrix, axis=1))

# Create the Laplacian matrix of G as a numpy array
laplacian_matrix = degree_matrix - adj_matrix

print(laplacian_matrix)
```

```python
from scipy.sparse.linalg import eigsh

# Convert the matrix to a supported type
laplacian_matrix = laplacian_matrix.astype(float)

m = 25  # number of eigenvectors to calculate
eigenvalues, eigenvectors = eigsh(laplacian_matrix, m, which='SM')

print('Eigenvalues:', eigenvalues)
print('Eigenvectors:\n', eigenvectors)
```

```python
pos = nx.kamada_kawai_layout(G)  # calculate the positions of the nodes in a regular grid formation
nx.draw(G, pos=pos, with_labels=False, node_color=eigenvectors[:, 0], cmap='coolwarm')
plt.show()
```

```python
# pos = nx.spring_layout(G)  # calculate the positions of the nodes

for i in range(m):
    nx.draw(G, pos=pos, with_labels=False, node_color=eigenvectors[:, i], cmap='coolwarm')
    plt.title(f'm={i}')
    plt.show()
```
