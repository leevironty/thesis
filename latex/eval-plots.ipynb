```python
def opt_weight_squared_error(ws) -> float:
    # print(ws)
    # raise Exception
    exploded = pd.DataFrame.from_records(ws.ravel())
    scaling = exploded.max(axis=1)
    exploded = exploded / scaling.values.reshape(-1, 1)
    pred_w = exploded.mean(axis=0)
    return ((exploded - pred_w) ** 2).values.mean()
```
